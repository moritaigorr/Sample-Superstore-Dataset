### Exploratory Data Analysis — Superstore Sales

#### Objetivo
O objetivo deste projeto é realizar uma análise exploratória de dados
sobre vendas de uma rede varejista, buscando identificar padrões,
tendências e possíveis oportunidades de melhoria no desempenho
de vendas e lucro.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,5)
```

#### Entendimento dos Dados


```python
df = pd.read_csv("../data/SampleSuperstore.csv")
#df.info() # Check the structure of the DataFrame 
df.shape # Check the dimensions of the DataFrame
#df.describe() # Get summary statistics for numerical columns
#df.head() # Display the first few rows of the DataFrame
```

#### Limpeza de Dados

Foram realizadas verificações de valores ausentes e duplicados
para garantir a qualidade da análise.

```python
df.isnull().sum() # Check for missing values
df.duplicated().sum() # Check for duplicate rows
df.drop_duplicates(inplace=True) # Remove duplicate rows if any
```

#### Como as vendas se distribuem?


```python
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title("Distribuição das Vendas")
plt.show()
```

#### Qual é o total de vendas por Categoria?

```python
df.groupby('Category')['Sales'].sum().sort_values().plot(kind='barh')
plt.title("Total de Vendas por Categoria")
plt.xlabel("Vendas")
plt.show()
```

#### Qual será as quantidade de VENDAS por REGIÃO?

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Vendas por Região")
plt.ylabel("Vendas")
plt.show()
```

#### Insights

- A maior parte das vendas está concentrada na categoria Technology.
- Algumas categorias apresentam alto volume de vendas, mas lucro baixo,
  indicando possíveis custos elevados.
- Existe uma tendência de crescimento das vendas ao longo do tempo,
  com picos em determinados períodos do ano.
- A região West apresenta o melhor desempenho em vendas.

#### Conclusão

A análise exploratória permitiu identificar padrões importantes no desempenho
de vendas e lucro da empresa. Esses insights podem auxiliar na tomada de decisão
estratégica, como otimização de estoque, foco em categorias mais lucrativas
e planejamento regional.
